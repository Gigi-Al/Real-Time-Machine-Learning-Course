```python
import pandas as pd

#Loading the data from csv file into pandas dataframe for better visualization
df = pd.read_csv ("Housing.csv")
```

```python
df
```

```python
import torch
from torchvision import transforms

#Converting the dataframe to tensor 
y = df['price'].values
x = df[["area", "bedrooms", "bathrooms", "stories", "parking"]].values
```

```python
import numpy as np

x = np.array(x)
y = np.array(y)
y = np.expand_dims(y, axis=1)
y.shape
```

```python
from sklearn import preprocessing

scaler_x = preprocessing.StandardScaler().fit(x)
scaler_y = preprocessing.StandardScaler().fit(y)
```

```python
x_scaled = scaler_x.transform(x)
y_scaled = scaler_y.transform(y)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split (x_scaled, y_scaled, test_size=0.2, random_state=42)
```

```python
X_train_t = torch.tensor(X_train, dtype=torch.float32)
X_test_t = torch.tensor(X_test, dtype=torch.float32)

y_train_t = torch.tensor(y_train, dtype=torch.float32)
y_test_t = torch.tensor(y_test, dtype=torch.float32)
```

```python
import torch.nn as nn
import torch.nn.functional as F

class Net(nn.Module):
    def __init__(self):
      super(Net, self).__init__()
      # First fully connected layer
      self.fc1 = nn.Linear(5, 8)
      # Second fully connected layer that outputs our 10 labels
      self.fc2 = nn.Linear(8, 1)

    def forward(self, x):
      # Pass data through fc1
      x = self.fc1(x)
      x = F.tanh(x)
      # Pass data through fc2
      x = self.fc2(x)
      output = F.tanh(x)
      return output
```

```python
def training_loop(n_epochs, optimizer, model, loss_fn, input_train, y_train, input_valid, y_valid):
    for epoch in range(1, n_epochs+1):
        predicted = model(input_train)
        loss_train = loss_fn (predicted, y_train)
        
        predicted_val = model (input_valid)
        loss_valid = loss_fn(predicted_val, y_valid)

        optimizer.zero_grad()
        loss_train.backward()
        optimizer.step()

        if epoch %10 == 0:
            print(f"Epoch {epoch}, Training loss {loss_train.item():.4f}, Validation loss {loss_valid.item():.4f}")
```

```python
model1 = Net()
optimizer = torch.optim.SGD(model1.parameters(), lr=0.01)
training_loop(200, optimizer, model1, nn.MSELoss(), X_train_t, y_train_t, X_test_t, y_test_t)
```

```python
class Net2(nn.Module):
    def __init__(self):
      super(Net2, self).__init__()
      # First fully connected layer
      self.fc1 = nn.Linear(5, 10)
      self.fc2 = nn.Linear(10, 8)
      self.fc3 = nn.Linear(8, 4)
      self.fc4 = nn.Linear(4, 1)

    def forward(self, x):
      x = self.fc1(x)
      x = F.tanh(x)
      x = self.fc2(x)
      x = F.tanh(x)
      x = self.fc3(x)
      x = F.tanh(x)
      x = self.fc4(x)
      output = F.tanh(x)
      return output
```

```python
model2 = Net2()
optimizer = torch.optim.SGD(model2.parameters(), lr=0.01)
training_loop(200, optimizer, model2, nn.MSELoss(), X_train_t, y_train_t, X_test_t, y_test_t)
```

## Problem 2

```python
class Net3(nn.Module):
    def __init__(self):
      super(Net3, self).__init__()
      # First fully connected layer
      self.fc1 = nn.Linear(32*32*3, 512)
      self.fc2 = nn.Linear(512, 10)
 

    def forward(self, x):
      x = self.fc1(x)
      x = F.tanh(x)
      x = self.fc2(x)
      output = F.log_softmax(x, dim=1)
      return output
```

```python
from torchvision import datasets

transform = transforms.Compose(
    [transforms.ToTensor(),
     transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))])

data_path = "./data"
cifar10 = datasets.CIFAR10(data_path, train=True, download=False, transform=transform)
cifar10_val = datasets.CIFAR10(data_path, train=False, download=False, transform=transform)

train_loader = torch.utils.data.DataLoader(cifar10, batch_size=64, shuffle=True)
val_loader = torch.utils.data.DataLoader(cifar10_val, batch_size=64, shuffle=False)
```

```python
def training_loop2(n_epochs, optimizer, model, loss_fn, input_train, input_valid):
    for epoch in range(1, n_epochs+1):
        total = 0 
        correct = 0
        for imgs, labels in input_train:
            batch_size = imgs.shape[0]
            outputs = model(imgs.view(batch_size, -1))
            loss = loss_fn(outputs, labels)
            _, predicted = torch.max(outputs, dim=1)
            total += labels.shape[0]
            correct += int((predicted==labels).sum()) 
            train_acc = correct/total

            optimizer.zero_grad()
            loss.backward()
            optimizer.step()

        total = 0
        correct = 0
        with torch.no_grad():
            for imgs, labels in input_valid:
                batch_size = imgs.shape[0]
                outputs = model(imgs.view(batch_size, -1))
                loss_valid = loss_fn(outputs, labels)
                _, predicted = torch.max(outputs, dim=1)
                total += labels.shape[0]
                correct += int((predicted==labels).sum()) 
                acc = correct/total

        if epoch %1 == 0:
            print(f"Epoch {epoch}, Training loss {loss.item():.4f}, Training accuracy: {train_acc}, Validation loss {loss_valid.item():.4f}, Validation accuracy: {acc}")
```

```python
model3 = Net3()
optimizer = torch.optim.SGD(model3.parameters(), lr=0.01)
training_loop2(200, optimizer, model3, nn.NLLLoss(), train_loader, val_loader)
```

```python
model4 = Net4()
optimizer = torch.optim.SGD(model4.parameters(), lr=0.01)
training_loop2(200, optimizer, model4, nn.NLLLoss(), train_loader, val_loader)
```
